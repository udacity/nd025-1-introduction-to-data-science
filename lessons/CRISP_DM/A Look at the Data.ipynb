```python
# Create a pandas Series with counts of each Salary value
salary_counts = df['Salary'].value_counts()

# Create a pandas Series with counts of each Salary Expectation value
expected_salary_counts = df['ExpectedSalary'].value_counts()

# Combine the two Series into a DataFrame for plotting
combined_counts = pd.DataFrame({
    'Salary': salary_counts,
    'Salary Expectation': expected_salary_counts
}).fillna(0)  # Fill NaN values with 0 for missing counts

# Get the top 10 values for each column
top_salary = combined_counts['Salary'].nlargest(10)
top_expectation = combined_counts['Salary Expectation'].nlargest(10)

# Combine the top 10 values into a new DataFrame
top_combined = pd.DataFrame({
    'Salary': top_salary,
    'Salary Expectation': top_expectation
}).fillna(0)  # Fill NaN values with 0 for missing counts

# Filter out rows where both "Salary" and "Salary Expectation" are 0
top_combined = top_combined[(top_combined['Salary'] != 0) & (top_combined['Salary Expectation'] != 0)]

# Plotting
ax = top_combined.plot(kind='bar', figsize=(10, 6), width=0.8)

# Adding values on top of the bars
for i in range(len(top_combined)):
    ax.text(i - 0.2, top_combined['Salary'].iloc[i] + 0.5, int(top_combined['Salary'].iloc[i]), ha='center', va='bottom')
    ax.text(i + 0.2, top_combined['Salary Expectation'].iloc[i] + 0.5, int(top_combined['Salary Expectation'].iloc[i]), ha='center', va='bottom')

# Format x-axis to show values in thousands
ax.set_xticklabels([f'{int(x / 1000)}K' for x in top_combined.index], rotation=0)

# Keep y-axis counts in original values
plt.title('Top 10 Counts of Salary and Salary Expectation')
plt.xlabel('Salary Values (in thousands)')
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.legend(title='Category')
plt.grid(axis='y')  # Add grid lines for better readability
plt.show()
```

```python
# Count occurrences for each column
counts = {
    'QuestionsInteresting': df['QuestionsInteresting'].value_counts(),
    'QuestionsConfusing': df['QuestionsConfusing'].value_counts(),
    'InterestedAnswers': df['InterestedAnswers'].value_counts(),
    'SurveyLong': df['SurveyLong'].value_counts()
}

# Create a figure for the pie charts with a larger size
fig, axs = plt.subplots(2, 2, figsize=(16, 12))  # Increased figure size
fig.suptitle('Survey Responses Pie Charts', fontsize=16)

# Create pie charts for each question
for ax, (question, count) in zip(axs.flatten(), counts.items()):
    wedges, texts, autotexts = ax.pie(
        count,
        labels=None,  # Disable labels on the pie chart to avoid overlap
        autopct='%2.2f%%',  # Display all percentages
        startangle=140,  # Adjust start angle for better distribution
        radius=0.8,
        pctdistance=0.9,  # Increase distance of percentage text from the center
        textprops=dict(color="black", fontsize=10)  # Customize text properties for percentages
    )

    # Create a legend instead of labels
    ax.legend(count.index, title=question, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    
    # Customize the appearance of the percentage text
    for text in autotexts:
        text.set_color('white')  # Set percentage text color
        text.set_fontsize(10)    # Set percentage text size

    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    ax.set_title(question)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to make room for the title
plt.show()


```
